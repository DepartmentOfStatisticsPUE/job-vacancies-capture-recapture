## Simulation study 1

We generate counts according for $N_A$ and $N_B$ according to the following probabilities
\begin{equation}
    \begin{aligned}
    p_{11} & = (1-\alpha)p_1 p_2, \\ 
    p_{10} & = \alpha p_1 + (1-\alpha) p_1 (1-p_2) = p_1(\alpha + (1-\alpha)(1-p_2)), \\
    p_{01} & =  \alpha (1-p_1) + (1-\alpha)(1-p_1)p_2 =(1-p_1)(\alpha - (1-\alpha)p_2), \\
    p_{00} & = (1-\alpha)(1-p_1)(1-p_2),
    \end{aligned}
    \label{eq-model1-probs}
\end{equation}

where $p_{1a}=p_{1b}=p_{1}$ and we use $p_{2a} \neq p_{2b}$ which are motivated by the empirical data from the survey and online data.

```julia
Random.seed!(123)
n_sims = 5000
N_a = 50_000
N_b = 20_000
x_a = sample_from_n(N_a, 0.05, 0.15, 0.05, n_sims)
x_b = sample_from_n(N_b, 0.05, 0.15, 0.15, n_sims)
sim_data = vcat(x_a[1:3,:], x_b[1:3,:])
```

```julia
sim_result = map(cr_averse_optimize, eachcol(sim_data));
```

Summary of results

```julia
sim_results_mat = hcat(sim_result...)
```

```julia
## naive
N_a_naive = map(x -> (x[1]+x[2])*(x[1]+x[3])/x[1], eachcol(sim_data))
N_b_naive = map(x -> (x[4]+x[5])*(x[4]+x[6])/x[4], eachcol(sim_data))
```

```julia
truevals = [N_a, N_b, 0.05, 0.15, 0.05, 0.15]
expvals = mean(sim_results_mat, dims = 2)[:,1]
bias =  (expvals - truevals)./expvals.*100
stdvals = std(sim_results_mat, dims = 2)[:,1]

sim_1_tab = DataFrame(:expval => expvals, :bias => bias, :cv => stdvals ./expvals .*100)

latexify(sim_1_tab; env=:table, latex=false, adjustment = :r, fmt = "%.4f")
```

```julia
naivevals = [mean(N_a_naive), mean(N_b_naive)]
biasvals = (naivevals .- truevals[1:2])./naivevals*100
naive_std_vals = [std(N_a_naive), std(N_b_naive)]

sim_1_tab2 = DataFrame(:expval => naivevals, :bias => biasvals, :cv => naive_std_vals ./naivevals .*100)

latexify(sim_1_tab2; env=:table, latex=false, adjustment = :r, fmt = "%.4f")
```

Coverage rates for the confidence intervals

```julia
N_means = mean(sim_results_mat[1:2,:], dims = 2)
N_stds = std(sim_results_mat[1:2,:], dims = 2)
q_95 = quantile(Normal(),1-0.05/2)
N_cis = hcat(N_means .- q_95 .*N_stds, N_means .+ q_95 .*N_stds)
## standard way
standard_coverage = sum( (sim_results_mat[1:2,:] .> N_cis[:,1]) .& (sim_results_mat[1:2,:] .< N_cis[:,2]), dims = 2) ./ n_sims

## proposed
x0s = mean(sim_data, dims = 2)
x0s = [sum(x0s[1:3]), sum(x0s[4:6])]
Ĉ = exp.(q_95 .* sqrt.(log.(1 .+ N_stds.^2 ./ (N_means .- x0s).^2)))
N_cis_prop = hcat(x0s .+ (N_means .- x0s) ./ Ĉ, x0s .+ (N_means .- x0s) .* Ĉ)

proposed_coverage = sum( (sim_results_mat[1:2,:] .> N_cis_prop[:,1]) .& (sim_results_mat[1:2,:] .< N_cis_prop[:,2]), dims = 2) ./ n_sims

ci_results = vcat( hcat(N_cis, standard_coverage), hcat(N_cis_prop, proposed_coverage))

latexify(ci_results; env=:table, latex=false, adjustment = :r, fmt = "%.4f")
```

## Simulation study 2

In simulation study 2 we verify robustness of the proposed estimator to voliation of main assumption i.e. $p_{1a} \neq p_{1b}$ and how it compares to the bias and MSE of the naive estimator. 


We generate counts according for $N_A$ and $N_B$ according to the following probabilities


\begin{equation}
    \begin{aligned}
    p_{11} & = (1-\alpha)p_1 p_2, \\ 
    p_{10} & = \alpha p_1 + (1-\alpha) p_1 (1-p_2) = p_1(\alpha + (1-\alpha)(1-p_2)), \\
    p_{01} & =  \alpha (1-p_1) + (1-\alpha)(1-p_1)p_2 =(1-p_1)(\alpha - (1-\alpha)p_2), \\
    p_{00} & = (1-\alpha)(1-p_1)(1-p_2),
    \end{aligned}
    \label{eq-model1-probs}
\end{equation}

where $p_{1a}=p_{1b}=p_{1}$ and we use $p_{2a} \neq p_{2b}$ which are motivated by the empirical data from the survey and online data.



```julia
Random.seed!(123)
N_a = 50_000
N_b = 20_000
p_1a=0.01:0.01:0.35
p_1b=0.15
## for loop
sim_2_table = zeros(length(p_1a)*length(p_1b), 9);
```

Takes around an hour

```julia
k = 1
@showprogress for a in p_1a
    Random.seed!(k)
    x_a = sample_from_n(N_a, 0.05, a,    0.05, n_sims)
    x_b = sample_from_n(N_b, 0.05, 0.15, 0.15, n_sims)
    sim_data_1 = vcat(x_a[1:3,:], x_b[1:3,:])
    sim_result_1 = map(cr_averse_optimize, eachcol(sim_data_1))
    sim_result_1_mat = hcat(sim_result_1...)
    
    N_a_naive = map(x -> (x[1]+x[2])*(x[1]+x[3])/x[1], eachcol(sim_data_1))
    N_b_naive = map(x -> (x[4]+x[5])*(x[4]+x[6])/x[4], eachcol(sim_data_1))
    
    @inbounds    sim_2_table[k, 1] = a 
    @inbounds    sim_2_table[k, 2:3] = mean(sim_result_1_mat, dims = 2)[1:2]
    @inbounds    sim_2_table[k, 4] = mean(N_a_naive)
    @inbounds    sim_2_table[k, 5] = mean(N_b_naive)
    @inbounds    sim_2_table[k, 6:7] = std(sim_result_1_mat, dims = 2)[1:2]
    @inbounds    sim_2_table[k, 8] = std(N_a_naive)
    @inbounds    sim_2_table[k, 9] = std(N_b_naive)
    k += 1
end    
```

Bias plot

```julia
p1 = plot( sim_2_table[:,1], (sim_2_table[:,2] .- N_a) ./1000, 
        color = "red", lw = 2,
        legend = :bottomright, label = L"N_A",
        ylabel = "Bias (in thousands)", xlabel = "Values of p_1A", 
        xticks = 0:0.05:0.35,
        ylim = (-70,70))
p1 = plot!(sim_2_table[:,1], (sim_2_table[:,3] .- N_b) ./1000, color = "blue", 
        label = L"N_B", lw = 2,
        legend = :bottomright)
p1 = plot!(sim_2_table[:,1], (sim_2_table[:,4] .- N_a) ./1000, color = "black", 
    label = L"N_{A,naive}",lw = 2,
        legend = :bottomright)
p1 = plot!(sim_2_table[:,1], (sim_2_table[:,5] .- N_b) ./1000, color = "gray", 
        label =L"N_{B,naive}", lw = 2,
        legend = :bottomright)
savefig(p1, "../figures/sim-2-s1-p1.pdf")
p1
```

MSE plot

```julia
p1a = plot( sim_2_table[:,1], 
        sqrt.( (sim_2_table[:,2] .- N_a).^2 .+ sim_2_table[:, 6].^2) ./ 1000, 
        color = "red", lw = 2,
        legend = :topright, label = L"N_A",
        ylabel = "RMSE (in thousands)", xlabel = "Values of p_1A", 
        xticks = 0:0.05:0.35,
        ylim = (0,70))
p1a = plot!(sim_2_table[:,1], 
        sqrt.( (sim_2_table[:,3] .- N_b).^2 .+ sim_2_table[:, 7].^2) ./1000, color = "blue", 
        label = L"N_B", lw = 2,
        legend = :topright)
p1a = plot!(sim_2_table[:,1], 
        sqrt.((sim_2_table[:,4] .- N_a).^2 .+ sim_2_table[:, 8].^2 ) ./1000, color = "black", 
        label = L"N_{A,naive}",lw = 2,
        legend = :topright)
p1a = plot!(sim_2_table[:,1], 
        sqrt.((sim_2_table[:,5] .- N_b).^2 .+ sim_2_table[:,9].^2) ./1000, color = "gray", 
        label =L"N_{B,naive}", lw = 2,
        legend = :topright)
savefig(p1a, "../figures/sim-2-s1-p1-rmse.pdf")
p1a
```

```julia
p2 = plot(sim_2_table[:,1], 
         (sim_2_table[:,2] .+ sim_2_table[:,3] .- (N_a + N_b)) ./1000,
        color = "red", lw = 2,
        legend = :bottomright, label = L"N",
        ylabel = "Bias (in thousands)", xlabel = "Values of p_1A", 
        xticks = 0:0.05:0.35,
        ylim = (-70,70))
p2 = plot!(sim_2_table[:,1], 
        (sim_2_table[:,4] .+ sim_2_table[:,5] .- (N_a + N_b)) ./1000,
        color = "blue", 
        label = L"N_{naive}", lw = 2,
        legend = :bottomright)
savefig(p2, "../figures/sim-2-s1-p2.pdf")
p2
```

```julia
p2a = plot(sim_2_table[:,1], 
          sqrt.((sim_2_table[:,2] .+ sim_2_table[:,3] .- (N_a + N_b)).^2 .+ 
                (sim_2_table[:,6]).^2 .+ sim_2_table[:,7].^2) ./1000,
        color = "red", lw = 2,
        legend = :topright, label = L"N",
        ylabel = "RMSE (in thousands)", xlabel = "Values of p_1A", 
        xticks = 0:0.05:0.35,
        ylim = (0,70))
p2a = plot!(sim_2_table[:,1], 
        sqrt.((sim_2_table[:,4] .+ sim_2_table[:,5] .- (N_a + N_b)) .^2 .+ 
               sim_2_table[:, 8].^2 .+ sim_2_table[:, 9].^2) ./1000,
        color = "blue", 
        label = L"N_{naive}", lw = 2,
        legend = :topright)
savefig(p2a, "../figures/sim-2-s1-p2-rmse.pdf")
p2a
```
